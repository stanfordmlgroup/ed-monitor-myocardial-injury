```python
from sklearn.metrics import roc_curve, auc, precision_recall_curve, roc_auc_score


def sig(path_1, path_2):
    df_1 = pd.read_csv(path_1)
    df_2 = pd.read_csv(path_2)
    df_1 = df_1.sort_values("patient_id")
    df_2 = df_2.sort_values("patient_id")
    
    y_actual = np.array(df_1["actual"].tolist())
    y_preds_1 = np.array(df_1["preds"].tolist())
    y_preds_2 = np.array(df_2["preds"].tolist())
    
#     print(len(y_actual))
#     print(len(y_preds_1))
#     fpr_keras, tpr_keras, thresholds_keras = roc_curve(y_actual, y_preds_1)
#     auc_keras = auc(fpr_keras, tpr_keras)
#     print(auc_keras)
    
    permutation_test_between_clfs(y_actual, y_preds_1, y_preds_2, nsamples=1000)
```
