## 12-Lead + Monitoring Waveforms

### Create Concatenated Waveforms

We will create a new embeddings file containing the concatenation of the single lead + 12-lead

#### 256-dim

```python
df_embeddings_summary_12 = pd.read_csv("/deep/group/ed-monitor/patient_data_v9/waveforms/12-lead/transformer-256/embeddings_summary.csv")
print(df_embeddings_summary_12.shape)
df_embeddings_summary = pd.read_csv("/deep/group/ed-monitor/patient_data_v9/waveforms/15sec-500hz-1norm-10wpp/II/transformer-64/embeddings_summary.csv")
print(df_embeddings_summary.shape)
```

```python
embeddings_12 = np.load("/deep/group/ed-monitor/patient_data_v9/waveforms/12-lead/transformer-256/embeddings.dat.npy")
print(embeddings_12.shape)
embeddings = np.load("/deep/group/ed-monitor/patient_data_v9/waveforms/15sec-500hz-1norm-10wpp/II/transformer-64/embeddings.dat.npy")
print(embeddings.shape)
```

```python
# 12-leads dataset are a subset of the full dataset so let's do a left join

monitoring_csn_to_index = {}
for i, row in df_embeddings_summary.iterrows():
    monitoring_csn_to_index[row["record_name"]] = i

print(f"monitoring_csn_to_index = {len(monitoring_csn_to_index)}")

output_embeddings = []
for i, row in df_embeddings_summary_12.iterrows():
    ind = monitoring_csn_to_index[row["record_name"]]
    output_row = []
    output_row.extend(embeddings_12[i])
    output_row.extend(embeddings[ind])
    output_embeddings.append(output_row)

output_embeddings = np.array(output_embeddings)
print(output_embeddings.shape)
```

```python
# Save files to new output folder 

from pathlib import Path
Path("/deep/group/ed-monitor/patient_data_v9/waveforms/12-lead/transformer-256-with-monitoring").mkdir(parents=True, exist_ok=True)
```

```python
df_embeddings_summary_12.to_csv(f"/deep/group/ed-monitor/patient_data_v9/waveforms/12-lead/transformer-256-with-monitoring/embeddings_summary.csv", index=False)
np.save(f"/deep/group/ed-monitor/patient_data_v9/waveforms/12-lead/transformer-256-with-monitoring/embeddings.dat.npy", output_embeddings)
```
